### Dictionary + Expanding On Some Code Segments

```python
# Creating a dictionary with information about the album "Title"
title_album = {
    "title": "Title",
    "artist": "Meghan Trainor",
    "year": 2015,
    "genre": ["Pop", "Dance Pop"],
    "tracks": {
        1: "Like I'm gonna Lose You",
        2: "The Best Part (Interlude)",
        3: "Title",
        4: "All About That Bass",
        5: "No Good For You",
        6: "Credit",
        7: "Walkashame",
        8: "What If I",
        9: "Mr. Almost",
        10: "Bang Dem Sticks",
        11: "Close Your Eyes",
        12: "My Selfish Heart",
        13: "Dear Future Husband",
        14: "Lips Are Movin"
    }
}


# Printing the dictionary
print(title_album)
```

```python
# Retrieve value from dictionary with key
print(title_album.get('tracks'))
# or
print(title_album['tracks'])
```

```python
# Retrieve value from a dictionary inside a dictionary
print(title_album.get('tracks')[1])
# or
print(title_album['tracks'][13])
```

```python
# adding a value with a new key
title_album["producer"] = set(['Meghan Trainor', 'Kevin Kadish', 'Chris Gelbuda', 'Meghan Trainor', 'Justin Trainor'])

# The set at the beginning to the list of producers is so that when printed, there will only be one "Meghan Trainor" printed instead of the two that are presented in the list

# Printing the dictionary
print(title_album)
```

**Updating Database**

```python
# Adding a an key-value pair to an existing key 
title_album["tracks"].update({15: "3am"})



# updating the dictionary
print(title_album)
```
