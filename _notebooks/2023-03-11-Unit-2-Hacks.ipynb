- Data compression can reduce the size (number of bits) of transmitted or stored data

- The amount of size reduction from compression depends on both the amount of redudancy in the original data representation and the compression algorithm applied.

- **Lossless Data** compression algorithms can usually reduce the number of bits sotred or transmitted while guaranteeing complete reconstruction of the original data.

- **Lossy Data** compression algorithms can siggnificantly reduce the number if bits stored ore transmitted but only allow reconstruction of and approximation of the original data.
