**Data About Iris Petals:**

```python
import seaborn as sns

iris_data = sns.load_dataset('iris')

print(iris_data)
```

**Mean of the Iris Data:**

```python
import numpy as np


mean = np.mean(iris_data)

print(mean)
```

**Maximum of Iris Data:**

```python
import numpy as np


maximum = np.max(iris_data)

print(maximum)
```

**Minimum of Iris Data:**

```python
import numpy as np


minimum = np.min(iris_data)

print(minimum)
```
